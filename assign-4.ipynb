```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings 
warnings.filterwarnings('ignore')
```

```python
mean=100
std_deviation=4
datasize=100000
```

creating a dataset with random numbers from datasize.

```python
data=np.random.normal(mean,std_deviation,datasize)
```

Ploting a histogram to show normal distribution.

```python
# Create the plot
plt.figure(figsize=(10,6))
sns.histplot(data, bins=50, kde=True, color='blue', stat='density', edgecolor='black', alpha=0.7)
sns.kdeplot(data, color='orange', linewidth=2)

# Add title, labels, and gridlines
plt.title('Normal Distribution with Mean=100, Std Dev=4, Size=100,000', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True)

# Show the plot
plt.show()

```

The smooth curve ensures normal distribution across the dataset. As the datasize is large so we get to see a perfect bell shape with smooth curve.
